# Анализ данных для интернет-магазина "Стримчик"

## Описание исследований
****
Исследование проводится для интернет-магазина «Стримчик», который продаёт по всему миру компьютерные игры.
****
**Цель исследования:** Выявление закономерностей, определяющих успешность игр. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
****
**Описание данных:**
Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).
* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.
****
**Этапы исследования:**
1. Загрузка данных и первичное ознакомление с ними;
1. Предобработка данных для дальнейшего анализа;
1. Проведение исследовательского анализа, примерный список вопросов:
  * Сколько игр выпускалось в разные годы? Важны ли данные за все периоды?
  * Как менялись продажи по платформам? За какой характерный срок появляются новые и исчезают старые платформы?
  * Какие платформы лидируют по продажам, растут или падают?
  * Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков?
  * Что можно сказать о самых прибыльных жанрах? Выделяются ли жанры с высокими и низкими продажами?
1. Определить для пользователя каждого региона (NA, EU, JP):
  * Самые популярные платформы (топ-5)
  * Самые популярные жанры (топ-5)
  * Влияет ли рейтинг ESRB на продажи в отдельном регионе?
1. Проверка гипотез:
  * Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
  * Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
1. Сделать общий вывод, исходя из проведенного исследования, и дать свои рекомендации на 2017 год.

#### Вывод по `name` и `genre`
Тут мы видим, что пропуски в столбцах `name` и `genre` в одних и тех же строках, без названия и жанра данные игры не подойдут для целевого анализа. Но продажи достаточно значимы, поэтому заменим пропуски в названиях на `unknown_name` и в жанрах на `Misc`, так как это жанр, который можно перевести как «прочий жанр».

```python
# заменяем пропуски
df['name'] = df['name'].fillna('unknown_name')
df['genre'] = df['genre'].fillna('Misc')

# проверим замену
df[['name', 'genre']].isna().sum()
```

#### Вывод по `year_of_release`

Есть замечательная библиотека `re`, и мы сможем заменить пропуски там, где явно есть в названии год, но год в названии не всегда означает дату релиза. Также можно найти открытую базу данных для заполнения пропусков по типу википедии, фандом. Но пропуски лучше оставить как есть, если мы не будем тратить время для ручного заполнения.

```python
# создадим чистую выборку без пропусков для анализа
df_clean = df.dropna(subset=['year_of_release', 'critic_score', 'user_score', 'rating'])
df_clean.isna().sum()
```

```python
# обработка типов  данных для чистой выборки
df_clean = df_clean[df_clean['user_score'] != 'tbd'] # это тоже самое что и  Nan
df_clean['user_score'] = df_clean['user_score'].astype('float64')
df_clean['critic_score'] = df_clean['critic_score'].astype('int64')
df_clean['year_of_release'] = df_clean['year_of_release'].astype('int64')
df_clean.sample(5)
```

```python
df_clean.info()
```

```python
# приехало из дубликатов для сохранения последовательности (единственная пара игр с одинаковыми названиями)
df.loc[(df['name'] == 'Need for Speed: Most Wanted') & (df['year_of_release'] == 2012), 'name'] = 'Need for Speed: Most Wanted(2012)'
df[(df['name'] == 'Need for Speed: Most Wanted(2012)') & (df['year_of_release'] == 2012)]
```

```python
# Вычисляем средние значения critic_score для каждой игры по другим платформам
mean_critic_score = df.groupby('name')['critic_score'].transform(
    lambda x: x.mean() if len(x) > 1 else np.nan
)
# Заполняем пропуски в critic_score
df['critic_score'] = np.round(df['critic_score'].fillna(mean_critic_score)).astype('Int64')
```

```python
df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan
```

```python
df['user_score'] = df['user_score'].astype('float64')
```

```python
# Вычисляем средние значения user_score для каждой игры по другим платформам
mean_user_score = df.groupby('name')['user_score'].transform(lambda x: x.mean() if len(x) > 1 else np.nan)

# Заполняем пропуски в user_score
df['user_score'] = np.round(df['user_score'].fillna(mean_user_score), 1)
```

```python

# Вычисляем моду для rating для каждой игры по другим платформам
mode_rating = df.groupby('name')['rating'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)

# Заполняем пропуски в user_score
df['rating'] = df['rating'].fillna(mode_rating)
```

## Проверка гипотез:

### **Гипотеза №1:** Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

* **Нулевая гипотеза** - **Средние пользовательские рейтинги** платформ Xbox One и PC **равны**.
* **Альтернативная гипотеза** - **Средние пользовательские рейтинги** платформ Xbox One и PC **различаются.**

```python
# создадим выборки для проверки гипотезы
hippo_xbox = df_clean_actual[df_clean_actual['platform'] == 'XOne']
hippo_pc = df_clean_actual[df_clean_actual['platform'] == 'PC']
hippo_xbox.shape[0], hippo_pc.shape[0]
```

```python
print(hippo_xbox['user_score'].dtype)
print(hippo_pc['user_score'].dtype)
```

```python
alpha = 0.05 # устанавливаем уровень статистической значимости 5%

# выведем средние двух выборок
print('Средний пользовательский рейтинг Xbox One: ', np.round(np.mean(hippo_xbox['user_score']), 2))
print('Средний пользовательский рейтинг PC: ', np.round(np.mean(hippo_pc['user_score']), 2))

results = st.ttest_ind(
    hippo_xbox['user_score'].astype(float), # какая то ошибка при переноси из гугл колаба .astype(float)-починил
    hippo_pc['user_score'].astype(float)
)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
  print('Отвергаем нулевую гипотезу')
else:
  print('Не получилось отвергнуть нулевую гипотезу')
```

#### Вывод:
* При уровне значимости `alpha = 0.05` (стандартное значение) `p-value = 0.128` больше `alpha`. Нет оснований отвергнуть нулевую гипотезу.
* Средние пользовательские рейтинги Xbox One (6.58) и PC (6.34) не имеют статистически значимых различий.

Итог: Нет статистических доказательств того, что пользователи оценивают игры на Xbox One и PC по-разному. Разница в рейтингах незначима.

### **Гипотеза №2:** Средние пользовательские рейтинги жанров Action и Sports разные.

* **Нулевая гипотеза** - **Средние пользовательские рейтинги** жанров Action и Sports **равны**.
* **Альтернативная гипотеза** - **Средние пользовательские рейтинги** жанров Action и Sports **различаются.**

```python
# создадим выборки для проверки гипотезы
hippo_action = df_clean_actual[df_clean_actual['genre'] == 'Action']
hippo_sports = df_clean_actual[df_clean_actual['genre'] == 'Sports']
hippo_action.shape[0], hippo_sports.shape[0]
```

```python
alpha = 0.05 # устанавливаем уровень статистической значимости 5%

# выведем средние двух выборок
print('Средний пользовательский рейтинг жанра Action: ', np.round(np.mean(hippo_action['user_score']), 2))
print('Средний пользовательский рейтинг жанра Sports: ', np.round(np.mean(hippo_sports['user_score']), 2))

results = st.ttest_ind(
    hippo_action['user_score'].astype(float), # какая то ошибка при переноси из гугл колаба .astype(float)-починил
    hippo_sports['user_score'].astype(float)
)


print('p-value:', results.pvalue)

if results.pvalue < alpha:
  print('Отвергаем нулевую гипотезу')
else:
  print('Не получилось отвергнуть нулевую гипотезу')
```

#### Вывод:
* При уровне значимости `alpha = 0.05` (стандартное значение) `p-value ≈ 0` меньше `alpha`. Отвергаем нулевую гипотезу.
* Средние пользовательские рейтинги жанра Action (6.87) и жанра Sports (5.45) имеют статистически значимые различия.

Итог: Разница в рейтингах статистически значима и, вероятно, отражает реальные предпочтения пользователей.

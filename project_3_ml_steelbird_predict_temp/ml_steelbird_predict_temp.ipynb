```python
# –≤—ã—á–∏—Å–ª–∏–º –î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å_–Ω–∞–≥—Ä–µ–≤–∞
data_arc_new['–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å_–Ω–∞–≥—Ä–µ–≤–∞'] = (
    data_arc_new['–ö–æ–Ω–µ—Ü_–Ω–∞–≥—Ä–µ–≤–∞_–¥—É–≥–æ–π'] - data_arc_new['–ù–∞—á–∞–ª–æ_–Ω–∞–≥—Ä–µ–≤–∞_–¥—É–≥–æ–π']
).dt.total_seconds()
data_arc_new['–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å_–Ω–∞–≥—Ä–µ–≤–∞'].describe()
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –ø—Ä–æ—Å—Ç–æ—è
def calculate_downtime(group):
    group = group.copy()
    group = group.sort_values('–ù–∞—á–∞–ª–æ_–Ω–∞–≥—Ä–µ–≤–∞_–¥—É–≥–æ–π')
    
    # –í—Ä–µ–º—è –ø—Ä–æ—Å—Ç–æ—è = –Ω–∞—á–∞–ª–æ —Ç–µ–∫—É—â–µ–≥–æ –Ω–∞–≥—Ä–µ–≤–∞ - –∫–æ–Ω–µ—Ü –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –Ω–∞–≥—Ä–µ–≤–∞
    group['–í—Ä–µ–º—è_–ø—Ä–æ—Å—Ç–æ—è'] = group['–ù–∞—á–∞–ª–æ_–Ω–∞–≥—Ä–µ–≤–∞_–¥—É–≥–æ–π'] - group['–ö–æ–Ω–µ—Ü_–Ω–∞–≥—Ä–µ–≤–∞_–¥—É–≥–æ–π'].shift(1)
    
    # –î–ª—è –ø–µ—Ä–≤–æ–≥–æ –Ω–∞–≥—Ä–µ–≤–∞ –≤ –≥—Ä—É–ø–ø–µ –ø—Ä–æ—Å—Ç–æ—è –Ω–µ—Ç (–∏–ª–∏ NaN)
    return group
```

```python
# –ü—Ä–∏–º–µ–Ω—è–µ–º –∫ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ
data_arc_with_downtime = data_arc_new.groupby('key').apply(calculate_downtime, include_groups=False).reset_index(drop=True)

# –ü–µ—Ä–µ–≤–æ–¥–∏–º –≤—Ä–µ–º—è –ø—Ä–æ—Å—Ç–æ—è –≤ —Å–µ–∫—É–Ω–¥—ã
data_arc_new['–í—Ä–µ–º—è_–ø—Ä–æ—Å—Ç–æ—è'] = data_arc_with_downtime['–í—Ä–µ–º—è_–ø—Ä–æ—Å—Ç–æ—è'].dt.total_seconds()

data_arc_new.head(10)
```

```python
# –ü–æ—Å—á–∏—Ç–∞–µ–º –ø–æ–ª–Ω—É—é –º–æ—â–Ω–æ—Å—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ (—Ç–µ–æ—Ä–µ–º–∞ –ø–∏—Ñ–∞–≥–æ—Ä–∞)
data_arc_new['–ü–æ–ª–Ω–∞—è_–º–æ—â–Ω–æ—Å—Ç—å'] = np.sqrt(
    data_arc_new['–†–µ–∞–∫—Ç–∏–≤–Ω–∞—è_–º–æ—â–Ω–æ—Å—Ç—å'] ** 2 + data_arc_new['–ê–∫—Ç–∏–≤–Ω–∞—è_–º–æ—â–Ω–æ—Å—Ç—å'] ** 2)

#  –ü–æ—Å—á–∏—Ç–∞–µ–º —ç–Ω–µ—Ä–≥–∏—é –∑–∞—Ç—Ä–∞—á–µ–Ω–Ω—É—é –Ω–∞ –Ω–∞–≥—Ä–µ–≤ —ç—Ç–æ –ø–æ–ª–Ω–∞—è_–º–æ—â–Ω–æ—Å—Ç—å * –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å_–Ω–∞–≥—Ä–µ–≤–∞
data_arc_new['–≠–Ω–µ—Ä–≥–∏—è'] = data_arc_new['–ü–æ–ª–Ω–∞—è_–º–æ—â–Ω–æ—Å—Ç—å'] * data_arc_new['–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å_–Ω–∞–≥—Ä–µ–≤–∞']

data_arc_new[['–ü–æ–ª–Ω–∞—è_–º–æ—â–Ω–æ—Å—Ç—å', '–≠–Ω–µ—Ä–≥–∏—è']].describe()
```

* –ë–æ–ª—å—à–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç–∏ –≤ –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö, –ø–æ—Å–ª–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –Ω—É–∂–Ω–æ –∑–∞–Ω—è—Ç—å—Å—è —Ñ–∏—á-–∏–Ω–∂–∏–Ω–∏—Ä–∏–Ω–≥–æ–º.

```python
def get_eda(df):

    for col in df.select_dtypes(include='object').columns.to_list():

        display(df[col].value_counts())

        plt.figure(figsize=(10, 5))
        ax = sns.countplot(x=col, data=df)
        plt.title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ {col}')
        plt.xlabel(col)
        plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

        for p in ax.patches:
            height = p.get_height()
            percent = 100 * height / len(df)
            ax.annotate(f'{percent:.1f}%',
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', fontsize=10)


        plt.xticks(rotation=30, ha='right')
        plt.tight_layout()
        plt.show()

    num_col = df.select_dtypes(exclude='object').columns.to_list()
    if num_col:
        display(df[num_col].describe().T)

    for col in num_col:

        if df[col].nunique() <= 20:
            plt.figure(figsize=(10, 5))
            ax = sns.countplot(x=col, data=df)
            plt.title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ {col}')
            plt.xlabel(col)
            plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')

            for p in ax.patches:
                height = p.get_height()
                percent = 100 * height / len(df)
                ax.annotate(f'{percent:.1f}%',
                          (p.get_x() + p.get_width() / 2., height),
                          ha='center', va='bottom', fontsize=10)
            plt.show()

        else:
            plt.figure(figsize=(10, 5))
            plt.subplot(2, 1, 1)
            sns.histplot(data=df, x=col, kde=True, bins=30)
            plt.title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ {col}')
            plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è')
            plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
            plt.grid(True)

            plt.subplot(2, 1, 2)
            sns.boxplot(data=df, x=col, orient='h')
            plt.title(f"Boxplot: {col}")
            plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è')
            plt.grid(True)

            plt.tight_layout(pad=2.0)
            plt.show()
```

## –û–±—É—á–µ–Ω–∏–µ –±–∞–∑–æ–≤–æ–π –º–æ–¥–µ–ª–∏ (LinReg)

```python
def get_model_lr(X, y, plot_residuals=True):

    # –°–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω: –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ + –º–æ–¥–µ–ª—å
    model = Pipeline([
        ("scaler", StandardScaler()),
        ("regressor", LinearRegression())
    ])


    # –ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è
    cv = KFold(n_splits=4, shuffle=True, random_state=RANDOM_STATE)
    neg_mae_scores = cross_val_score(model, X, y, cv=cv, scoring='neg_mean_absolute_error')

    # –í—ã–≤–æ–¥
    print("MAE –ø–æ —Ñ–æ–ª–¥–∞–º:", -neg_mae_scores)
    print("–°—Ä–µ–¥–Ω—è—è MAE:", -neg_mae_scores.mean())

    y_pred = cross_val_predict(model, X, y, cv=cv)
    
    # –ê–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤
    residuals = y - y_pred
    
    if plot_residuals:
        plt.figure(figsize=(15, 10))
        
        # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤
        plt.subplot(2, 2, 1)
        sns.histplot(residuals, kde=True)
        plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤')
        plt.xlabel('–û—à–∏–±–∫–∞ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è')
        plt.axvline(x=0, color='r', linestyle='--')
        
        # –ì—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞—Ç–∫–æ–≤ vs –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
        plt.subplot(2, 2, 2)
        plt.scatter(y_pred, residuals, alpha=0.5)
        plt.axhline(y=0, color='r', linestyle='--')
        plt.title('–û—Å—Ç–∞—Ç–∫–∏ vs –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è')
        plt.xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è')
        plt.ylabel('–û—Å—Ç–∞—Ç–∫–∏')
        
        # QQ-–≥—Ä–∞—Ñ–∏–∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏
        plt.subplot(2, 2, 3)
        stats.probplot(residuals, dist="norm", plot=plt)
        plt.title('QQ-–≥—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞—Ç–∫–æ–≤')
        
        # –ì—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
        plt.subplot(2, 2, 4)
        sns.ecdfplot(residuals)
        plt.title('–ö—É–º—É–ª—è—Ç–∏–≤–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤')
        plt.xlabel('–û—à–∏–±–∫–∞ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è')
        
        plt.tight_layout()
        plt.show()
    
    # –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –æ—Å—Ç–∞—Ç–∫–æ–≤
    print("\n–ê–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤:")
    print(f"–°—Ä–µ–¥–Ω–µ–µ –æ—Å—Ç–∞—Ç–∫–æ–≤: {residuals.mean():.4f}")
    print(f"–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤: {residuals.std():.4f}")
    print(f"–ê—Å–∏–º–º–µ—Ç—Ä–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤: {stats.skew(residuals):.4f}")
    
    result = permutation_importance(model.fit(X, y), X, y, scoring='neg_mean_absolute_error')

    feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': result['importances_mean']})
    feature_importance = feature_importance.sort_values('Importance', ascending=True)
    sns.set_style('white')
    feature_importance.plot(x='Feature', y='Importance', kind='barh')

    plt.title('Feature Importance')
    plt.xlabel('Value')
    plt.show()
    
    return model.fit(X, y)
```

## –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ CatBoost

–í–æ–∑–Ω–∏–∫–ª–∏ –ø—Ä–æ–±–ª–µ–º–∏ —Å–æ–≤–º–µ—Å—Ç–∏–º–æ—Å—Ç–∏ CatBoost –∏ GridSearchCV, –ø—Ä–∏—à–ª–æ—Å—å –ø–æ–¥–±–æ—Ä –¥–µ–ª–∞—Ç—å –≤ —Ä—É—á–Ω—É—é...

```python
def tune_catboost(X, y, param_grid, n_splits=4, random_state=RANDOM_STATE):
    # –ü–æ–ª—É—á–∞–µ–º –∏–Ω–¥–µ–∫—Å—ã –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
    cat_features = X.select_dtypes(include=['object', 'category']).columns
    cat_idx = [X.columns.get_loc(c) for c in cat_features]

    best_score = np.inf
    best_params = None
    best_model = None

    cv = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    for depth, lr, l2, it in product(param_grid['depth'], param_grid['learning_rate'],
                                     param_grid['l2_leaf_reg'], param_grid['iterations']):
        fold_scores = []

        for train_idx, val_idx in cv.split(X):
            X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
            y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

            train_pool = Pool(X_train.values, y_train.values, cat_features=cat_idx)
            val_pool = Pool(X_val.values, y_val.values, cat_features=cat_idx)

            model = CatBoostRegressor(
                depth=depth,
                learning_rate=lr,
                l2_leaf_reg=l2,
                iterations=it,
                random_state=random_state,
                verbose=0
            )
            model.fit(train_pool, eval_set=val_pool, verbose=False)
            y_pred = model.predict(val_pool)
            fold_scores.append(mean_absolute_error(y_val, y_pred))

        mean_score = np.mean(fold_scores)

        if mean_score < best_score:
            best_score = mean_score
            best_params = {"depth": depth, "learning_rate": lr, "l2_leaf_reg": l2, "iterations": it}
            best_model = model

    print("–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:", best_params)
    print("MAE:", best_score)
    
    # –ê–Ω–∞–ª–∏–∑ –≤–∞–∂–Ω–æ—Å—Ç–∏
    result = permutation_importance(best_model, X.values, y.values, scoring='neg_mean_absolute_error')

    feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': result['importances_mean']})
    feature_importance = feature_importance.sort_values('Importance', ascending=True)
    sns.set_style('white')
    feature_importance.plot(x='Feature', y='Importance', kind='barh')

    plt.title('Feature Importance')
    plt.xlabel('Value')
    plt.show()
    
    return best_model
```

```python
# –ø–æ–¥–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
param_grid = {
    "depth": [4, 6, 8],
    "learning_rate": [0.01, 0.05, 0.1],
    "l2_leaf_reg": [1, 3, 5],
    "iterations": [200, 400, 600]
}
```

```python
model_boost = tune_catboost(X_train, y_train, param_grid)
```
